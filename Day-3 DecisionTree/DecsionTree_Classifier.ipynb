# Decision Tree Classifier Model

# 1. Importing the requires libraries and Dataset

```python
from sklearn.datasets import load_iris      #Importing the Iris Species Dataset form Kaggle
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
data
```

# Splitting the Data into Training set Testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=99)
```

# Model Selection

```python
clf = DecisionTreeClassifier(random_state=1)
```

# Model Training

```python
clf.fit(X_train, y_train)   #To train the mode we use .fit
```

# Predicting the Outcome from the Trained Model

```python
y_pred = clf.predict(X_test)
```

# Model Evaluation

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy Score: {accuracy}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': range(1, 10, 1),
    'min_samples_leaf': range(1, 20, 2),
    'min_samples_split': range(2, 20, 2),
    'criterion': ["entropy", "gini"]
}

tree = DecisionTreeClassifier(random_state=1)

grid_search = GridSearchCV(estimator=tree, param_grid=param_grid, cv=5, verbose=True)

grid_search.fit(X_train, y_train)

print("Best Accuracy:", grid_search.best_score_)
print(grid_search.best_estimator_)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

tree_clf = grid_search.best_estimator_

plt.figure(figsize=(18, 15))
plot_tree(tree_clf, filled=True, feature_names=data.feature_names, class_names=list(data.target_names))
plt.show()
```
